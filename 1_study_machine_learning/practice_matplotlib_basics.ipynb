```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
           s=100*features[3], c=iris.target, cmap='viridis') # 4개 특성을 한번에 볼 수 있다 !
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
# plt.plot 과 plt.scatter의 비교:
# plt.plot은 한 특성을 복제
# plt.scatter는 개별적인 size, color 특성을 입혀야 함
# 따라서 대량의 데이터에서는 plt.plot이 효율적 (개별 특성 고려안하는 경우)
```

```python
# errorbar
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
            ecolor='lightgray', elinewidth=3, capsize=0)
```

```python
from sklearn.gaussian_process import GaussianProcess

model = lambda x: x*np.sin(x)
xdata = np.array([1,3,5,6,8])
ydata = model(xdata)

gp = GaussianProcess(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1,
                    random_start=100)

gp.fit(xdata[:, np.newaxis], ydata)
xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2*np.sqrt(MSE)
```

```python
# continuous errorbar: 1000개의 샘플에 1000개의 오차 찍는 것보다 효율적.

plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')

plt.fill_between(xfit, yfit-dyfit, yfit+dyfit,
                color='gray', alpha=0.2) # x값, 하한, 상한값을 전달하면 그 사이를 채워줌.

plt.xlim(0, 10)
```

```python
# histogram
plt.style.use('seaborn-white')

data = np.random.randn(1000)

plt.hist(data)
```

```python
# 특성
plt.hist(data, bins=30, normed=True, alpha=0.5,
        histtype='stepfilled', color='steelblue',
        edgecolor='none')
```

```python
# 여러 히스토그램 비교는 histtype='stepfilled'와 투명도 alpha를 결합하여 !
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 0.5, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

```python
# 디스플레이 말고 계산만 하고 싶다면 np.histogram

counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```
